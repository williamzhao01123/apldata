Week 4


*   Simulate trading outcomes given the historical data of AMPL,
*   Create visualization to search for the optimal trading strategy.



Data Import

```
import pandas as pd
urlAMPL = 'https://raw.githubusercontent.com/williamzhao01123/crypto-lab/master/Data/Ampleforth.csv'
AMPL = pd.read_csv(urlAMPL)
numrows = AMPL.shape[0]
AMPL
```

Simulation

I consider the behavior of holding AMPL to be equivalent to holding shares of market cap of AMPL. No matter how AMPL rebases the currency, the total market cap will not change, and the percent of market share of holders will not change. Rebasing does not change the percentage of market share.

```
def simulation(mclow, mchigh):
  cash = 1000
  holding = 0

  for index in range(numrows):
    mcmkt = AMPL['market_cap'][index]
    if mcmkt < mclow:
      holding = holding + cash/mcmkt
      cash = 0
    if mcmkt > mchigh:
      cash = cash + holding*mcmkt
      holding = 0
  outcome = cash + holding * mcmkt
  return(outcome)

simulation(1000, 20000000000)
```

Generating Results with Different Strategies

```
mclow = 1000000
mchigh = 2000000
high = []
low = []
outcome = []

while mchigh < 1000000001:
  while mclow < mchigh:
    high.append(mchigh)
    low.append(mclow)
    outcome.append(simulation(mclow, mchigh))
    mclow = mclow + 5000000
  mchigh = mchigh + 5000000
  mclow = 1000000

data = {'mclow':low, 'mchigh':high, 'outcome': outcome}
df = pd.DataFrame(data)
print(df)
```

Generating Visualization

```
import matplotlib.pyplot as plt
df['sign'] = np.sign(df['outcome'])

# Creating figure
fig = plt.figure(figsize=(20,15))
ax = plt.axes(projection ="3d")
colors = {1:'red', -1:'green'} 

# Creating plot
ax.scatter3D(df['mclow'], df['mchigh'], df['outcome'], c=df['sign'].map(colors), s=1 )
plt.title("Outcomes of all trading strategies")

# Setting Labels
ax.set_xlabel('buy-in market cap', fontweight ='bold')
ax.set_ylabel('sell-out market cap', fontweight ='bold')
ax.set_zlabel('trading outcome', fontweight ='bold')

# show plot
plt.show()
```
